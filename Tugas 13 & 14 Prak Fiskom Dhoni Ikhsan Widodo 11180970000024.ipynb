```python
import math
import numpy as np
from euler import *
from run_kut4 import *
from printSoln import *
import matplotlib.pyplot as plt
```

**Using h = 1.0**

```python
def F(x,y):
     #Equation based on problem
    F = np.zeros(2) 
    F[0] = y[1] 
    F[1] = -math.sin(y[0]) 
    return F

# x as taw
# y as theta
x = 0.0 # start value of x
xStop = 10.0 #end value of x
y = np.array([1.0,0.0]) #Initial value of y
h = 1.0 #Precision
freq = 50 #for printSoln function to print out the results

A,B = euler(F,x,y,xStop,h) #Euler function
X,Y = integrate(F,x,y,xStop,h) #Rungekutta Function

# the Exact result
yExact = np.cos(A)

print('These graphs represent the level of accuracy for Euler and Runge Kutta methods')
print('When h =',h)

# plotting graphs to see an accuracy of all methods 
plt.plot(A,B[:,0],'o-b',X,yExact,'.-g')
plt.title('Euler with Exact',size=15)
plt.grid()
plt.xlabel('x (Taw)')
plt.ylabel('y (Theta)')
plt.legend(('Euler','Exact'))
plt.show()

plt.plot(X,Y[:,0],'-r',X,yExact,'.-g')
plt.title('Rungekutta with Exact',size=15)
plt.grid()
plt.xlabel('x (Taw)')
plt.ylabel('y (Theta)')
plt.legend(('Runge kutta','Exact'))
plt.show()

plt.plot(A,B[:,0],'o-b',X,Y[:,0],'-r',X,yExact,'.-g')
plt.title('Comparison of each method to the Exact value',size=15)
plt.grid()
plt.xlabel('x (Taw)')
plt.ylabel('y (Theta)')
plt.legend(('Euler','Runge kutta','Exact'))
plt.show()
```
