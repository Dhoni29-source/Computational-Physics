```python
#Mencari nilai C = A + B, lalu mencari determinan C menggunakan metode Sarrus, C = matrix orde 3

A = [[1,2,3],
    [4,5,6],
    [7,8,9]]

B = [[1,0,0],
    [0,1,0],
    [0,0,1]]

print("A =")
print("  ", A[0][0], " ", A[0][1], " ", A[0][2])
print("  ", A[1][0], " ", A[1][1], " ", A[1][2])
print("  ", A[2][0], " ", A[2][1], " ", A[2][2])
      
print("B =")
print("  ", B[0][0], " ", B[0][1], " ", B[0][2])
print("  ", B[1][0], " ", B[1][1], " ", B[1][2])
print("  ", B[2][0], " ", B[2][1], " ", B[2][2])
    
C = [[0,0,0],[0,0,0],[0,0,0]]
C[0][0] = A[0][0] + B[0][0]
C[0][1] = A[0][1] + B[0][1]
C[0][2] = A[0][2] + B[0][2]
C[1][0] = A[1][0] + B[1][0]
C[1][1] = A[1][1] + B[1][1]
C[1][2] = A[1][2] + B[1][2]
C[2][0] = A[2][0] + B[2][0]
C[2][1] = A[2][1] + B[2][1]
C[2][2] = A[2][2] + B[2][2]

print("C = A + B")
print("  ", C[0][0], " ", C[0][1], " ", C[0][2])
print("  ", C[1][0], " ", C[1][1], " ", C[1][2])
print("  ", C[2][0], " ", C[2][1], " ", C[2][2])

#mencari nilai determinan C menggunakan metode Sarrus
print("det(C) = ", ((C[0][0] * C[1][1] * C[2][2]) + (C[0][1] * C[1][2] * C[2][0]) + 
      (C[0][2] * C[1][0] * C[2][1])) - ((C[2][0] * C[1][1] * C[0][2]) + 
      (C[2][1] * C[1][2] * C[0][0]) + (C[2][2] * C[1][0] * C[0][1])))
```

```python
#penjumlahan matrix orde 3
D = [[3,4,5],
    [7,8,9],
    [8,6,9]]

E = [[5,6,7],
    [7,8,9],
    [4,5,6]] 

print("D =")
print("  ", D[0][0], " ", D[0][1], " ", D[0][2])
print("  ", D[1][0], " ", D[1][1], " ", D[1][2])
print("  ", D[2][0], " ", D[2][1], " ", D[2][2])
      
print("E =")
print("  ", E[0][0], " ", E[0][1], " ", E[0][2])
print("  ", E[1][0], " ", E[1][1], " ", E[1][2])
print("  ", E[2][0], " ", E[2][1], " ", E[2][2])
    
F = [[0,0,0],[0,0,0],[0,0,0]]
F[0][0] = D[0][0] + E[0][0]
F[0][1] = D[0][1] + E[0][1]
F[0][2] = D[0][2] + E[0][2]
F[1][0] = D[1][0] + E[1][0]
F[1][1] = D[1][1] + E[1][1]
F[1][2] = D[1][2] + E[1][2]
F[2][0] = D[2][0] + E[2][0]
F[2][1] = D[2][1] + E[2][1]
F[2][2] = D[2][2] + E[2][2]

print("F = D + E")
print("  ", F[0][0], " ", F[0][1], " ", F[0][2])
print("  ", F[1][0], " ", F[1][1], " ", F[1][2])
print("  ", F[2][0], " ", F[2][1], " ", F[2][2])
```

```python
### Mencari nilai matrix invers orde 3
#Menggunakan fungsi loops (while)
#fungsi ini akan berulang terus, jadi bisa lanjut ke soal2 berikutnya tentang pencarian invers
a=1
a=1
while a==1:
 #Baris pemisah.....
     print(" ")
     print(" ")

     print("Masukkan Nilai Matriks nya woi")
     #Input soal nya
     a=int(input(" Nilai a = "))
     b=int(input(" Nilai b = "))
     c=int(input(" Nilai c = "))
     d=int(input(" Nilai d = "))
     e=int(input(" Nilai e = "))
     f=int(input(" Nilai f = "))
     g=int(input(" Nilai g = "))
     h=int(input(" Nilai h = "))
     i=int(input(" Nilai i = "))

    #Mencari Determinan
     detA=(a*e*i)+(b*f*g)+(c*d*h)-(g*e*c)-(h*f*a)-(i*d*b)
     print("\nMatriks A")
     print("| ",a," ",b," ",c," |")
     print("| ",d," ",e," ",f," |")
     print("| ",g," ",h," ",i," |")
     print("  ")   
     print(" Det A = ",detA)

    #Mencari Adjoin
     a11=(e*i)-(h*f)
     a12=(d*i)-(g*f)
     a13=(d*h)-(g*e)
     a21=(b*i)-(h*c)
     a22=(a*i)-(g*c)
     a23=(a*h)-(g*b)
     a31=(b*f)-(e*c)
     a32=(a*f)-(d*c)
     a33=(a*e)-(d*b)

     #Rumus Adjoin 
     print("  ")
     print('-----Adjoin-----')
     print("A11 = + | ",e,"",f," |=",a11," A12 =  - | ",d,"",f," |=",a12," A13 = + | ",d,"",e," |=",a13)
     print("        | ",h,"",i," |                  | ",g,"",i," |                 | ",g,"",h," |")

     print("\nA21 = - | ",b,"",c," |=",a21," A22 = + | ",a,"",c," |=",a22," A23 = - | ",a,"",b," |=",a23)
     print("          | ",h,"",i," |                 | ",g,"",i," |                 | ",g,"",h," |")

     print("\nA31 = + | ",b,"",c," |=",a31," A32 = - | ",a,"",c," |=",a32," A33 = + | ",a,"",b," |=",a33)
     print("          | ",e,"",f," |                 | ",d,"",f," |                 | ",d,"",e," |")

    #Menampilkan Adjoin
     print("------------------------------------------------------------------------------")
     print("Adj = | ",a11*(1),"",a12*(-1),"",a13*(1)," |")
     print("      | ",a21*(-1),"",a22*(1),"",a23*(-1)," |")
     print("      | ",a31*(1),"",a32*(-1),"",a33*(1)," |")
     print("")
     print("Adjoin di Transpose menjadi = ")
     print("Adjoin Transpose = | ",a11*(1),'',a21*(-1),'',a31*(1),' |')
     print('                   | ',a12*(-1),'',a22*(1),'',a32*(-1),' |')
     print('                   | ',a13*(1),'',a23*(-1),'',a33*(1),' |')
     print("-------------------------------------------------------------------------------")

     #Rumus invers
     inva11=(1/detA*(a11*(1)))
     inva12=(1/detA*(a12*(-1)))
     inva13=(1/detA*(a13*(1)))
     inva21=(1/detA*(a21*(-1)))
     inva22=(1/detA*(a22*(1)))
     inva23=(1/detA*(a23*(-1)))
     inva31=(1/detA*(a31*(1)))
     inva32=(1/detA*(a32*(-1)))
     inva33=(1/detA*(a33*(1)))

     #Menampilkan hasil matrix invers dari metode Adjoin
     print("INVERS =")   
     print("\nA-1 = 1/Det A x Adj A")
     print('    = 1/',detA,"| ",a11,'',a21,'',a31," |")
     print('                | ',a12,'',a22,'',a32,' |')
     print('                | ',a13,'',a23,'',a33,' |')

     print('\n    = | ',inva11,'',inva21,'',inva31,' |')
     print('        | ',inva12,'',inva22,'',inva32,' |')
     print('        | ',inva13,'',inva23,'',inva33,' |')
     print(" ")
     print("Semangaattttt!!!!!")
     print("Jangan lupa bilang makasih sama programnya")
```
