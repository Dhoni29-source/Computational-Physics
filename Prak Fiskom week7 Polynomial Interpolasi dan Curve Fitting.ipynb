**Polynomial Interpolation**
pake Newton methods untuk mencari angka yang hilang disekitar titik yg kontinue

```python
# Interpolasi polynomial using Newton methods
# Example 3.4 hlm 114
# Membandingkan yInterp dengan yExact 
import numpy as np
import math
from newtonPoly import *

xData = np.array([0.15,2.3,3.15,4.85,6.25,7.95])
yData = np.array([4.79867,4.49013,4.2243,3.47313,2.66674,1.51909])
a = coeffts(xData,yData)
print(" x    yInterp   yExact") 

print("-----------------------")
for x in np.arange(0.0,8.1,0.5): #Mulai dari 0.0 sampai 8.1 dengan selang 0.5
    y = evalPoly(a,xData,x)
    yExact = 4.8*math.cos(math.pi*x/20.0) #Sesuai keterangan soal f(x) = ......
    print('{:1.1f} {:9.5f} {:9.5f}'.format(x,y,yExact)) # {:1.1f} {:9.5f} {:9.5f} Menyatakan jarak dari depan dan jumlah angka di belakang koma
```

**Interpolation and Curve fitting**
Fitting menggunakan Metode Least square Interpolation sepanjang titik data yg kontinue

```python
## Fitting menggunakan polyFit
# least square methods
# Example 3.12 hlm 140
import numpy as np
from polyFit import *
from plotPoly import *
xData = np.array([-0.04,0.93,1.95,2.90,3.83,5.0, \
                  5.98,7.05,8.21,9.08,10.09])
yData = np.array([-8.66,-6.44,-4.36,-3.27,-0.88,0.87, \
                  3.31,4.63,6.19,7.4,8.85])
while True:
    try:
        m = eval(input("\nDegree of polynomial ==> "))
        coeff = polyFit(xData,yData,m)
        print("Coefficients are:\n",coeff)
        print("Std. deviation =",stdDev(coeff,xData,yData))
    except SyntaxError: break
print(plotPoly(xData,yData,coeff))
```

```python
import numpy as np
from rational import *
from neville import *
from newtonPoly import *
import matplotlib.pyplot as plt
import numpy as np
import math
from newtonPoly import *
from plotPoly import *
from polyFit import *

m = eval(input("\nDegree of polynomial ==> "))
xData = np.array([-2.0,1.0,4.0,-1.0,3.0,-4.0])
yData = np.array([-1.0,2.0,59.0,4.0,24.0,-53.0])
coeff = polyFit(xData,yData,m)
a = coeffts(xData,yData)
print("data x    data y") 
print("------------------")
for x in np.arange(-5.0,5.0,1.0): 
    y = evalPoly(a,xData,x)
    print('{:3.1f} {:12.5f}'.format(x,y))  
print(plotPoly(xData,yData,coeff))

x = np.arange(-5.0,5.0,1.0)
n = len(x)
y = np.zeros((n,2))
for i in range(n):
    y[i,0] = rational(xData,yData,x[i])
    y[i,1] = neville(xData,yData,x[i])
plt.plot(xData,yData,'o',x,y[:,0],'-',x,y[:,1],'--')
plt.xlabel('x')
plt.legend(('Data','Rational','Neville'),loc = 0)
plt.show()
```
