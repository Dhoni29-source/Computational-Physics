**Using Central different from finite different methods to find acceleration values based on 0, 5, 10 to 180 degress (radian)**

```python
import numpy as np
import math
import matplotlib.pyplot as plt

## Convert to international unit
# 5000 rev/min = 523.59878 rad/sec
# 90 mm = 0.09 m

# Declare function and equation
# Function based on problem
def f(x):
    return 0.09*(math.cos(x) + math.sqrt(2.5**2 - math.sin(x)**2)) 

# Equation for Central different based on problem state
def dff2(x):
    return (f(x+h) - 2*f(x) + f(x-h))*523.59878**2/h**2 

degrees = np.arange(0,37*math.pi/36,math.pi/36) #Degree and distance parameters for Looping to find acceleration 
h = 0.0005
Accel = np.zeros(len(degrees)) #Spot to fill acceleration values


# Looping to find acceleration values based on degrees parameter
for i in range (len(Accel)): 
    Accel[i] = dff2(degrees[i])

# Print the result of acceleration 
print('We can get accelerations when degrees:\n')
print('degrees \t  Acceleration(m/s^2)')
for y in range (len(Accel)):
    print('When',round((180/(math.pi))*degrees[y]),'=\t',Accel[y])

# Making an acceleration graphic
plt.plot(degrees,Accel,'--.b')
plt.title('Acceleration graphic', size=15)
plt.xlabel('radian (π)', size=12)
plt.ylabel('Acceleration (m/s^2)', size=12)
plt.grid()
plt.show()
```
