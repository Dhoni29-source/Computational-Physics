```python
#cell for problem 3
import numpy as np
import matplotlib.pyplot as plt
def neville(xData,yData,x):
    m = len(xData) # number of data points
    y = yData.copy()
    for k in range(1,m):
        y[0:m-k] = ((x - xData[k:m])*y[0:m-k] + \
                    (xData[0:m-k] - x)*y[1:m-k+1])/(xData[0:m-k] - xData[k:m])
    return y[0]

xData = np.array([-2.0, 1.0, 4.0, -1.0, 3.0, -4.0])
yData = np.array([-1.0, 2.0, 59.0, 4.0, 24.0, -53.0])
x = np.arange(-4.0,len(xData),1.0)
n = len(x)
y = np.zeros((n,2))
for i in range(n):
    y[i,1] = neville(xData,yData,x[i])
    
plt.title('Fitting menggunakan Neville', size=15)
plt.plot(xData,yData,'o',x,y[:,1],'--')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(('Data','Neville'),loc = 0)
plt.show()
```

```python
# Code dgn menerapkan import modul lagrange
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, simplify, lambdify
from lagrange import *

x = Symbol('x')
y = simplify(lagrange(x,[-2.0,1.0,4.0,-1.0,3.0,-4.0],[-1.0,2.0,59.0,4.0,24.0,-53.0])) #contoh datanya: kiri untuk x dan kanan untuk y

x1 = np.linspace(-4, 4, 100)
y1 = lambdify(x, y)(x1)

fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.scatter([-2.0,1.0,4.0,-1.0,3.0,-4.0],[-1.0,2.0,59.0,4.0,24.0,-53.0], c = 'r') #Plot data x dan y menggunakan scatter 
plt.title('Fitting menggunakan Lagrange', size=15)
plt.grid()
plt.xlabel('x', size=12)
plt.ylabel('y', size=12)
plt.show()
```

```python
# Code lainnya untuk menyelesaikan problem 4
# kita dapat mencari standar deviasi dengan polyFit function 
# dan meplotting/fitting data nya dengan neville, lalu membandingkan nya dengan rational function 
# grafik nya sama dengan cell diatas
# Jadi, jawaban nomor 4 nya sudah valid dan dapat dibuktikan dengan menggunakan neville, rational ataupun least square

import numpy as np
from rational import *
from neville import *
from newtonPoly import *
from polyFit import *
from plotPoly import *
import matplotlib.pyplot as plt

# Data
x_data = np.array([0.5,1.0,1.5,2.0,2.5])
y_data = np.array([0.541,0.398,0.232,0.106,0.052])
        
# Plotting with neville and compare with rational line 
x = np.arange(0.5,3.0,0.5)
n = len(x)
y = np.zeros((n,2))
for i in range(n):
    y[i,0] = rational(x_data,y_data,x[i])
    y[i,1] = neville(x_data,y_data,x[i])
    
plt.plot(x_data,y_data,'o',x,y[:,0],'-',x,y[:,1],'--')
plt.title('Fitting data menggunakan Neville dan Fungsi Rasional ', size=15)
plt.xlabel('x', size=15)
plt.ylabel('y', size=15)
plt.grid()
plt.legend(('Data','Rational','Neville'),loc = 0)
plt.show()
```

```python
# Code dgn menerapkan import modul lagrange
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, simplify, lambdify
from lagrange import *

x = Symbol('x')
y = simplify(lagrange(x,[0.5,1.0,1.5,2.0,2.5],
                      [0.541,0.398,0.232,0.106,0.052])) #contoh datanya: kiri untuk x dan kanan untuk y

x1 = np.linspace(0.5, 2.5, 5)
y1 = lambdify(x, y)(x1)

fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.scatter([0.5,1.0,1.5,2.0,2.5],[0.541,0.398,0.232,0.106,0.052], c = 'r') #Plot data x dan y menggunakan scatter 
plt.title('Fitting menggunakan Lagrange', size=15)
plt.grid()
plt.xlabel('x', size=12)
plt.ylabel('y', size=12)
plt.show()
```
