```python
# example rootsearch
# Halaman 147
from rootsearch import *

def f(x): return x**3 - 10.0*x**2 + 5.0
x1 = 0.0; x2 = 1.0
for i in range(4):
    dx = (x2 - x1)/10.0
    x1,x2 = rootsearch(f,x1,x2,dx)
x = (x1 + x2)/2.0
print('x =', '{:6.4f}'.format(x))
```

```python
# Example rootsearch and bisection
# halaman 150
import math
from rootsearch import *
from bisection import *

def f(x): return x - math.tan(x)
a,b,dx = (0.0, 20.0, 0.01)
print("The roots are:")
while True:
    x1,x2 = rootsearch(f,a,b,dx)
    if x1 != None:
        a = x2
        root = bisection(f,x1,x2,1)
        if root != None: print(root)
    else:
        print("\nDone")
        break
```

```python
# example Newton-Raphson 
# halaman 159

# fungsi awal
def f(x): return x**4 - 6.4*x**3 + 6.45*x**2 + 20.538*x - 31.752

# hasil turunan
def df(x): return 4.0*x**3 - 19.2*x**2 + 12.9*x + 20.538

# proses pencarian akar fungsi
def newtonRaphson(x,tol=1.0e-9):
    for i in range(30):
        dx = -f(x)/df(x)
        x = x + dx
        if abs(dx) < tol: return x,i
    print('Too many iterations\n')

root,numIter = newtonRaphson(2.0)
print('Root =',root)
print('Number of iterations =',numIter)
```

```python
# example newton-Rapshon2 
# halaman 165

# misalkan x = x1, y = x2 dan z = x3,
# penulisan di python x1=x[0], x[2]=x[1], x[3]=x[2]

import numpy as np
import math
from newtonRaphson2 import *

def f(x):
    f = np.zeros(len(x))
    
    # masukan fungsi
    f[0] = math.sin(x[0]) + x[1]**2 + math.log(x[2]) - 7.0
    f[1] = 3.0*x[0] + 2.0**x[1] - x[2]**3 + 1.0
    f[2] = x[0] + x[1] + x[2] - 5.0
    return f

# mulai dari point (1,1,1)
x = np.array([1.0, 1.0, 1.0])
print('x =', newtonRaphson2(f,x))
```
